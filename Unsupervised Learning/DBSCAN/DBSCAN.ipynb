# **DBSCAN**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Generate synthetic data
# `make_blobs` generates a set of points grouped into blobs (clusters).
# - `n_samples=1000`: This generates 1000 data points.
# - `centers=4`: It will generate 4 centers (clusters) for the blobs.
# - `random_state=42`: Setting a fixed random state ensures that the result is reproducible (same every time the code is run).
X, _ = make_blobs(n_samples=1000, centers=4, random_state=42)

# Visualize Core Data
# `plt.figure(figsize=(10, 5))` sets the size of the figure (plot).
# - `figsize=(10, 5)` means a plot that is 10 units wide and 5 units tall.
plt.figure(figsize=(10, 5))

# Create the scatter plot
# `plt.scatter(X[:, 0], X[:, 1], c='black')`:
# - `X[:, 0]` refers to the first feature (or dimension) of the data (x-axis).
# - `X[:, 1]` refers to the second feature (y-axis).
# - `c='black'` sets the color of the points to black.
plt.scatter(X[:, 0], X[:, 1], c='black')

# Set the title of the plot
# `plt.title('Blobs Data')` adds a title to the plot for clarity.
plt.title('Blobs Data')

# Label the x-axis and y-axis for clarity.
# `plt.xlabel('Feature 1')` and `plt.ylabel('Feature 2')` add axis labels.
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Display the plot
# `plt.show()` renders the plot to the screen.
plt.show()
```

```python
from sklearn.cluster import DBSCAN

# DBSCAN clustering
# DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a clustering algorithm
# that groups together closely packed points, marking points in low-density regions as outliers.
# - `eps=0.6`: This is the maximum distance between two samples for them to be considered as in the same neighborhood. A smaller `eps` will lead to more outliers, while a larger one will create larger clusters.
# - `min_samples=6`: This specifies the minimum number of samples (data points) required to form a dense region (cluster).
# The algorithm uses these two parameters to identify core points, border points, and noise points.
dbscan = DBSCAN(eps=0.6, min_samples=6)

# Apply the DBSCAN algorithm to the data
# `fit_predict(X)` computes the DBSCAN clustering on the input data `X` and assigns a cluster label to each point.
# Points labeled as -1 are considered outliers (noise), while other points are assigned a cluster label (e.g., 0, 1, 2,...).
dbscan_labels = dbscan.fit_predict(X)

# Visualize DBSCAN clustering
# `plt.figure(figsize=(12, 5))` creates a plot with a width of 12 units and height of 5 units.
plt.figure(figsize=(12, 5))

# Create the scatter plot to visualize the clustering
# - `X[:, 0]`: First feature of the data (x-axis).
# - `X[:, 1]`: Second feature of the data (y-axis).
# - `c=dbscan_labels`: The color of each point corresponds to its cluster label, which is determined by the DBSCAN algorithm. Points marked as -1 (outliers) will have a different color.
# - `cmap='viridis'`: The color map for displaying the clusters, where different cluster labels will have distinct colors.
# - `alpha=0.5`: This makes the points semi-transparent, allowing overlapping points to be visible more clearly.
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis', alpha=0.5)

# Add a title and axis labels for the plot.
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Display the plot
plt.show()
```

```python
from sklearn.datasets import make_moons

# Generate synthetic data with two moon-shaped clusters
# `make_moons` generates a synthetic dataset with two interleaving half circles (moon-shaped clusters).
# - `n_samples=1000`: This creates 1000 data points.
# - `noise=0.06`: This introduces some noise into the data. A noise value of `0.06` means that 6% of the points may not fit the true moon shapes, making the dataset more realistic (i.e., data is not perfectly clustered).
X, _ = make_moons(n_samples=1000, noise=0.06)

# Visualize Core Data
# `plt.figure(figsize=(10, 5))` creates a plot with a width of 10 units and height of 5 units for better readability.
plt.figure(figsize=(10, 5))

# Create the scatter plot
# `plt.scatter(X[:, 0], X[:, 1], c='black')`:
# - `X[:, 0]` refers to the first feature (x-axis) of the data.
# - `X[:, 1]` refers to the second feature (y-axis) of the data.
# - `c='black'` colors all points black, as we are visualizing the raw data here, without any clustering or categorization applied.
plt.scatter(X[:, 0], X[:, 1], c='black')

# Add a title and axis labels to the plot.
# `plt.title('Moons Data')` adds a title that indicates this is moon-shaped synthetic data.
plt.title('Moons Data')

# Label the x-axis and y-axis for clarity.
# `plt.xlabel('Feature 1')` and `plt.ylabel('Feature 2')` make the plot's axes more understandable.
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Display the plot
# `plt.show()` renders the plot on the screen.
plt.show()
```

```python
# DBSCAN clustering
# DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a clustering algorithm
# that finds core samples in dense areas and groups them together. It can identify clusters of arbitrary shape
# and separate out noise (outliers).
# - `eps=0.2`: This is the maximum distance between two points for them to be considered neighbors. A smaller `eps` will lead to more points being considered outliers, while a larger `eps` will group more points into clusters.
# - `min_samples=5`: This specifies the minimum number of samples required in a neighborhood to form a cluster. A smaller value will create smaller clusters, while a larger value will require more points to form a cluster.
dbscan = DBSCAN(eps=0.2, min_samples=5)

# Apply DBSCAN to the moon-shaped data `X`
# `fit_predict(X)` fits the DBSCAN model on the data and assigns a cluster label to each point.
# Points that are considered noise (outliers) will be labeled as `-1`, and points that belong to clusters
# will be assigned unique integer labels (0, 1, 2, ...).
dbscan_labels = dbscan.fit_predict(X)

# Visualize DBSCAN clustering
# `plt.figure(figsize=(10, 5))` creates a figure with a width of 10 units and height of 5 units.
plt.figure(figsize=(10, 5))

# Scatter plot to visualize the DBSCAN clustering
# - `X[:, 0]`: The first feature (x-axis) of the data.
# - `X[:, 1]`: The second feature (y-axis) of the data.
# - `c=dbscan_labels`: Colors the points based on their DBSCAN cluster label. Points labeled as `-1` (outliers) will have a distinct color.
# - `cmap='viridis'`: Uses the "viridis" color map, which provides distinct colors for each cluster.
# - `alpha=0.5`: Sets the transparency of the points to 50%, making the visualization clearer when points overlap.
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis', alpha=0.5)

# Add a title and axis labels to the plot
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Display the plot
plt.show()
```
